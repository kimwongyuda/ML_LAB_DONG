# 단어집 만들기

```python
vocabulary = {}

with open('SMSSpamCollection',"r", encoding="utf-8") as file_handle:
    for line in file_handle:
        #print(line)
        #print()
        
        splits = line.split()
        label = splits[0]
        text = splits[1:]
        
        for word in text:
            lower = word.lower()
            if not lower in vocabulary:
                vocabulary[lower] = len(vocabulary)
                
```

# 단어 빈도 피처 만들기

```python
import numpy as np

features = []
with open('SMSSpamCollection',"r", encoding="utf-8") as file_handle:
    for line in file_handle:
        feature = np.zeros(len(vocabulary))
        
        splits = line.split()
        text = splits[1:]
        
        for word in text:
            lower = word.lower()
            feature[vocabulary[lower]] += 1

        feature = feature/sum(feature)
        
        print(feature)
        
        print()
        features.append(feature)
```

# 레이블 처리하기

```python
labels = []
with open('SMSSpamCollection',"r", encoding="utf-8") as file_handle:
    for line in file_handle:
        splits = line.split()
        label = splits[0]
        if label == 'spam':
            labels.append(1)
        else:
            labels.append(0)
```

# 사이킷런 이용해서 문서 데이터를 피처 벡터로 변환하기 (위에거 배제하고 새로운 가공 데이터)


```python
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

spam_header = 'spam\t'
no_spam_header = 'ham\t'
documents = [] # 각 문서 내용 넣기
labels = []

with open('SMSSpamCollection',"r", encoding="utf-8") as file_handle:
    for line in file_handle:
        
        if line.startswith(spam_header):
            labels.append(1)
            documents.append(line[len(spam_header):])
        elif line.startswith(no_spam_header):
            labels.append(0)
            documents.append(line[len(no_spam_header):])
            
vectorizer = CountVectorizer() # 단어 횟수 피처 만들기
term_counts = vectorizer.fit_transform(documents) # 문서에서 단어 횟수 센다.
#print(term_counts)
vocabulary = vectorizer.get_feature_names()
#print(vocabulary)
tf_transformer = TfidfTransformer(use_idf=False).fit(term_counts)
features = tf_transformer.transform(term_counts)
#print(features)

with open('processed.pickle', 'wb') as file_handle:
    pickle.dump((vocabulary, features, labels), file_handle)
```

# 로지스틱 회귀 이용한 스팸 분류(위의 피클 이용)

```python
import pickle
from sklearn.linear_model import LogisticRegression

with open('processed.pickle', 'rb') as file_handle:
    vocabulary, features, labels = pickle.load(file_handle)

total_number = len(labels)
middle_index = total_number//2
train_features = features[:middle_index,:]
train_labels = labels[:middle_index]
test_features = features[middle_index:,:]
test_labels = labels[middle_index:]

#features = 빈도! 빈도를 학습

classifier = LogisticRegression()
classifier.fit(train_features, train_labels)

print('train accurancy: %4.4f' % classifier.score(train_features, train_labels))
print('test accurancy: %4.4f' % classifier.score(test_features, test_labels))
```

# Weights about Each features

```python
weights = classifier.coef_[0,:]
pairs = []
print(weights)
```

```python
for index, value in enumerate(weights):
    print(index)
    print(value)
    pairs.append((abs(value), vocabulary[index]))
pairs.sort(key=lambda x: x[0], reverse=True)

for pair in pairs[:20]:
    print('score(weight) %4.4f word: %s' % pair)
```
